```python
import pandas as pd
import plotly.graph_objects as go

def plot_signal_rolling_mean(df: pd.DataFrame, signal_name: str, window_size: int) -> go.Figure:
    fig = go.Figure()

    # Add the original series in light blue
    fig.add_trace(
        go.Scatter(
            x=df["frame"], 
            y=df[signal_name], 
            mode="lines", 
            name="Original Series", 
            line=dict(color="lightblue")
        )
    )

    # Add the rolling mean in dark blue
    fig.add_trace(
        go.Scatter(
            x=df["frame"], 
            y=df[signal_name].rolling(window=window_size).mean(), 
            mode="lines", 
            name=f"Rolling Mean ({window_size})", 
            line=dict(color="darkblue")
        )
    )

    # Update layout for style
    fig.update_layout(
        template="plotly_white", 
        title="Original Series and Rolling Mean",
        xaxis_title="Frame", 
        yaxis_title="Value"
    )

    return fig
```
